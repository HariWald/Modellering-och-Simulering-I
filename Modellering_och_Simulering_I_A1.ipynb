30. Mean temp

```python
Month = np.arange(1,13)
Temp = np.array([-0.6,-0.5,1.9,6.,11.4,15.4,16.8,16.5,13.,9.1,4.5,1.1])

fig,ax = plt.subplots()
ax.plot(Month,Temp,'o')

a = np.mean(Temp)
b = max(Temp)/2 - min(Temp)

delta = -2.25

y = a + b*np.sin(np.pi/6*Month + delta)

ax.plot(Month,y)
```

34. Plot surfaces

```python
# For moving 3D plot:
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

x = y = np.linspace(-3,3,100)

X,Y = np.meshgrid(x,y)

Z = 0.2*np.cos(X) + Y*np.exp(-X**2-Y**2)

#print(np.shape(X),np.shape(Y),np.shape(Z))

fig = plt.figure()
ax[0] = fig.add_subplot(121, projection='3d')
surf = ax[0].plot_surface(X,Y,Z)
ax[1] = fig.add_subplot(122)
cont = ax[1].contour(X,Y,Z)
```

36. Load images from the net

```python
from PIL import Image
import requests

urlstr = 'https://upload.wikimedia.org/wikipedia/commons/a/ad/Boltzmann2.jpg'

img = Image.open(requests.get(urlstr, stream=True).raw)
imarray = np.array(img)

#print(img.format,img.size,img.mode)

bbox_props = dict(boxstyle="round", fc="w", ec="0.5",)

fig,ax = plt.subplots()
Im = ax.imshow(imarray,cmap='gray')
ax.text(0.1,0.99,'I think .. I found a constant',transform=ax.transAxes,color='magenta',backgroundcolor='white',bbox=bbox_props)
ax.text(0.85,0.75,'$\sigma$',transform=ax.transAxes,color='blue',fontsize=20)
ax.axis('off')
```
